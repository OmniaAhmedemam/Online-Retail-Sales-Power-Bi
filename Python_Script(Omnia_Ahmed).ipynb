```python
# Import nedded libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data
file_path='/content/drive/MyDrive/sales_data_sample.csv'
df=pd.read_csv(file_path,encoding='latin1')
# Check the head of the dataframe
df.head()
```

```python
# Calculate orders for each product line
orders_by_productline = df.groupby('productline')['ordernumber'].nunique()
plt.figure(figsize=(10, 6))
plt.bar(orders_by_productline.index, orders_by_productline.values)
plt.xlabel('Product Line')
plt.ylabel('Number of Orders')
plt.title('Number of Orders per Product Line')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Combine similar product
def product_description(product):
    if product in ['Classic Cars', 'Vintage Cars', 'Motorcycles']:
        return 'Vehicles'
    elif product in ['Planes', 'Trucks and Buses','Ships','Trains']:
        return 'Heavy Vehicles'
    else:
        return 'Other'
df['product_category']= df['productline'].apply(product_description)
```

```python
# Calculate orders for each product category
orders_by_category = df.groupby('product_category')['ordernumber'].nunique()
plt.figure(figsize=(10, 6))
plt.bar(orders_by_category.index, orders_by_category.values)
plt.xlabel('Product Category')
plt.ylabel('Number of  Orders')
plt.title('Number of Orders per Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# segment customers based on purchase frequency and total spending amount
df['total_price'] = df['quantityordered'] * df['priceeach']
customer_segmentation = df.groupby('customername').agg(
    {'total_price': lambda total_price: total_price.sum(),
     'ordernumber': lambda num: num.nunique()})
customer_segmentation.columns=['spending','frequency']
customer_segmentation.head(10)
```

```python
customer_segmentation['frequency_score']=pd.qcut(customer_segmentation['frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
customer_segmentation['spending_score']=pd.qcut(customer_segmentation['spending'],5,labels=[1,2,3,4,5])
customer_segmentation.head(10)
```

```python
customer_segmentation['score_code']=(customer_segmentation['frequency_score'].astype(str)+customer_segmentation['spending_score'].astype(str))
customer_segmentation.head(10)
```

```python
seg_map={
    r'[1-2][1-2]':'low_value',
    r'[1-2][3-4]':'medium_value',
    r'[1-2]5': 'medium_value',
    r'3[1-2]':'medium_value',
    r'33' : 'medium_value',
    r'[3-4][4-5]':'high_value',
    r'41':'medium_value'   ,
    r'51':'medium_value' ,
    r'[4-5][2-3]':'high_value',
    r'5[4-5]':'high_value'
       }
customer_segmentation['segment']=customer_segmentation['score_code'].replace(seg_map,regex=True)
customer_segmentation.head()
```

```python
customer_segmentation['segment'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=customer_segmentation, x='segment')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.title('Customer Segment Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df.merge(customer_segmentation[['segment']], left_on='customername', right_index=True, how='left')
df.head()
```

```python
# CLTV= average order value * purshase frequency * averge customer lifetime
# Calculate average order value for each customer
average_order_value = df.groupby('customername')['sales'].sum() / df.groupby('customername')['ordernumber'].nunique()
average_order_value
```

```python
# Calculate average frequency rate
average_frequency_rate = df['ordernumber'].nunique() / df['customername'].nunique()

print(average_frequency_rate)
```

```python
# Calculate the first and last purchase dates for each customer
customer_lifetime = df.groupby('customername')['orderdate'].agg(['min', 'max']).reset_index()
customer_lifetime.head()


```

```python
# Calculate the active period in years
customer_lifetime['ActivePeriodYears'] = (customer_lifetime['max'] - customer_lifetime['min']).dt.days / 365.25

```

```python
#Calculate the average active period
average_active_period= customer_lifetime['ActivePeriodYears'].mean()
# Calculate average customer lifetime
average_customer_lifetime = average_active_period / df['customername'].nunique()


```

```python
cltv=average_order_value*average_frequency_rate*average_customer_lifetime
cltv
```

```python
cltv_df=cltv.reset_index()
cltv_df.columns=['customername','cltv']
cltv_df.head(15)
```

```python
df = pd.merge(df, cltv_df[['customername', 'cltv']], on='customername', how='left')
df.head()
```
